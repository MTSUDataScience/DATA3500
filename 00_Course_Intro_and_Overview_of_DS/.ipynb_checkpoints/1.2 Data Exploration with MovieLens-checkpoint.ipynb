```python
### Run the following code to mount your google drive.
### Follow the link that appears and enter the authorization code provided.
### Change the path to your file in your google drive folder.

from google.colab import drive
drive.mount('/content/gdrive')

path_1 = 'gdrive/My Drive/Data 3500/Data/movielens/users.dat' 
path_2 = 'gdrive/My Drive/Data 3500/Data/movielens/ratings.dat' 
path_3 = 'gdrive/My Drive/Data 3500/Data/movielens/movies.dat' 
```

```python
import pandas as pd

unames = ['user_id', 'gender', 'age', 'occupation', 'zip']
users = pd.read_table(path_1, sep='::', header=None, names=unames, engine='python')

rnames = ['user_id', 'movie_id', 'rating', 'timestamp']
ratings = pd.read_table(path_2, sep='::', header=None, names=rnames, engine='python')

mnames = ['movie_id', 'title', 'genres']
movies = pd.read_table(path_3, sep='::', header=None, names=mnames, engine='python')

movies[:5]
```

```python
### Combine files to allow analysis across seperate files.
### Notice that occupation is coded as an integer. See README file for data descriptions.

data = pd.merge(pd.merge(ratings, users), movies)
data
```

```python
### Aggregate data by gender (user, movie, etc...) to get average rankings across groups.

mean_rating = data.pivot_table('rating', index='title', columns='gender', aggfunc='mean')
mean_rating[:5]
```

```python
### Consider only those movies with at least 250 ratings.
### Create an index of all titles with at least 250 ratings.
### Use the index to subset your main data.

ratings_by_title = data.groupby('title').size()
active_titles = ratings_by_title.index[ratings_by_title >= 250]
mean_rating = mean_rating.loc[active_titles]
mean_rating
```

```python
### Identify the top films by female viewers.

top_female_ratings = mean_rating.sort_values(by='F', ascending=False)
top_female_ratings[:10]
```

```python
### Create a measure of gender disagreement.
### Identify which movies have the biggest difference in ratings.

mean_rating['diff'] = mean_rating['M'] - mean_rating['F']
mean_rating.sort_values(by='diff', inplace=True)
mean_rating[:10]
```

```python
### Identify the top movies preferred by men relative to women.

mean_rating[::-1][:10]
```

```python
### Which movies create the most disagreement among viewers, regardless of gender?

rating_std_by_title = data.groupby('title')['rating'].std()
rating_std_by_title = rating_std_by_title.loc[active_titles]
rating_std_by_title.sort_values(ascending=False)[:10]
```
